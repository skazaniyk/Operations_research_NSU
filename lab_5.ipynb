```python
import numpy as np
```

```python
file = open('binpack.txt', 'r')
first_line = file.readline()
bin_size = bin_size = int(first_line.replace('bin size=', '').strip())
lines = file.read().split('\n')
file.close()
items = np.array([float(line) for line in lines])
```

# Online

Next-fin

```python
def next_fit(items, bin_capacity):
    bins = []
    current_bin = []

    for item in items:
        if sum(current_bin) + item <= bin_capacity + 1e-5:
            current_bin.append(item)
        else:
            bins.append(current_bin)
            current_bin = [item]

    bins.append(current_bin)
    return bins

bins = next_fit(items, bin_size)
print(len(bins))
```

```python
bins
```

First-fit

```python
def first_fit(items, bin_capacity):
    bins = []

    for item in items:
        placed = False
        for bin in bins:
            if sum(bin) + item <= bin_capacity + 1e-5:
                bin.append(item)
                placed = True
                break

        if not placed:
            bins.append([item])

    return bins

bins = first_fit(items, bin_size)
print(len(bins))
```

```python
bins
```

# Online

First-fit-decreasing

```python

def first_fit_decreasing(items, bin_capacity):
    sorted_items = np.sort(items)[::-1]
    return first_fit(sorted_items, bin_capacity)

bins = first_fit_decreasing(items, bin_size)
print(len(bins))
```

```python
bins
```

Modified First-Fit-Decreasing (MFFD)

```python
def modified_first_fit_decreasing(items, bin_capacity):
    sorted_items = np.sort(items)[::-1]
    bins = []

    for item in sorted_items:
        bin_found = False
        
        for bin in bins:
            if sum(bin) + item <= bin_capacity:
                bin.append(item)
                bin_found = True
                break
    
        if not bin_found:
            bins.append([item])

    return bins

bins = modified_first_fit_decreasing(items, bin_size)
print(len(bins))
```

Martello Lower Bound

```python
import math
```

```python
martello_lower_bound = math.ceil(np.sum(items)/bin_size)
print(martello_lower_bound)
```

Toth Lower Bound

```python
toth_lower_bound = math.ceil(np.max(np.cumsum(np.sort(items)[::-1]) / bin_size))
print(toth_lower_bound)
```
